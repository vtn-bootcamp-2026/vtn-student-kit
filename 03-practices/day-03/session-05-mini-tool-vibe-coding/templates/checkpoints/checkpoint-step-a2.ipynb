# Checkpoint A2: 3 Agent Profiles đã thiết kế

3 SOUL.md profiles đã tạo. `skill.json` đã điền. Tiếp tục từ **Bước B1: Knowledge Base**.

```python
# Hiển thị các profile có sẵn
import subprocess

try:
    result = subprocess.run(
        ["hermes", "profile", "list"],
        capture_output=True, text=True, timeout=10
    )
    print(result.stdout if result.stdout else result.stderr)
except FileNotFoundError:
    print("hermes CLI không tìm thấy. Kiểm tra lại cài đặt Hermes Agent.")
except Exception as e:
    print(f"Lỗi: {e}")
```

```python
# Hiển thị nội dung SOUL.md cho từng profile
import os

profiles = ["hr-manager", "it-support", "finance-analyst"]
base = os.path.expanduser("~/.hermes/profiles")

for profile_name in profiles:
    soul_path = os.path.join(base, profile_name, "SOUL.md")
    print(f"{'='*60}")
    print(f"Profile: {profile_name}")
    print(f"{'='*60}")
    if os.path.exists(soul_path):
        with open(soul_path, "r") as f:
            content = f.read()
        print(content[:500])
        if len(content) > 500:
            print(f"\n... (còn {len(content) - 500} ký tự)")
    else:
        print(f"Chưa tìm thấy: {soul_path}")
    print()
```

```python
# Hiển thị nội dung skill.json
import os
import json

skill_path = os.path.expanduser("~/vtn-session05/skill.json")
if os.path.exists(skill_path):
    with open(skill_path, "r") as f:
        data = json.load(f)
    print("skill.json:")
    print(json.dumps(data, indent=2, ensure_ascii=False))
else:
    print(f"Chưa tìm thấy skill.json tại: {skill_path}")
    print("Tạo file skill.json theo hướng dẫn trong lab.md.")
```
