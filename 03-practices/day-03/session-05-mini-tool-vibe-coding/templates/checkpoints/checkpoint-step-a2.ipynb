# Checkpoint A2: 3 Agent Profiles đã thiết kế

Notebook này sẽ tự động khởi tạo 3 profiles tác tử Hermes và sao chép các tệp cấu hình SOUL.md tương ứng từ thư mục templates. `skill.json` cũng sẽ được tự động chuẩn bị tại thư mục làm việc `output/vtn-session05`.

```python
import os
import subprocess
import shutil
import json

profiles = ["tri-thuc-noi-bo", "soan-thao-noi-dung", "checklist-van-hanh"]
base_profiles_dir = os.path.expanduser("~/.hermes/profiles")

# 1. Xác định thư mục templates nguồn chứa các file soul-*.md và skill.json
notebook_dir = os.getcwd()
possible_paths = [
    os.path.abspath(os.path.join(notebook_dir, "..")),
    os.path.abspath("."),
    os.path.abspath(".."),
    os.path.abspath("../../templates"),
]
template_dir = None
for p in possible_paths:
    if os.path.exists(os.path.join(p, "soul-tri-thuc-noi-bo.md")):
        template_dir = p
        break

if template_dir:
    print(f"📂 Thư mục templates nguồn: {template_dir}")
else:
    backup_dir = "/Users/shimazu/Documents/9. active/course_viettel_network/vtn-ai-builders-bootcamp-2026/03-practice/day-03/session-05-mini-tool-vibe-coding/templates"
    if os.path.exists(backup_dir):
        template_dir = backup_dir
        print(f"📂 Thư mục templates nguồn (dự phòng): {template_dir}")
    else:
        print("❌ LỖI: Không tìm thấy thư mục templates chứa các file soul-*.md.")

# 2. Khởi tạo các profile, copy file SOUL.md và config.yaml
for profile in profiles:
    profile_dir = os.path.join(base_profiles_dir, profile)
    soul_target = os.path.join(profile_dir, "SOUL.md")
    config_target = os.path.join(profile_dir, "config.yaml")
    
    if not os.path.exists(profile_dir):
        print(f"⚙️ Đang tạo profile '{profile}'...")
        try:
            res = subprocess.run(
                ["hermes", "profile", "create", profile, "--clone"],
                capture_output=True, text=True, timeout=15
            )
            if res.returncode == 0:
                print(f"✅ Đã tạo thành công profile: {profile}")
            else:
                print(f"❌ Lỗi khi tạo profile {profile}: {res.stderr.strip()}")
                os.makedirs(profile_dir, exist_ok=True)
        except Exception as e:
            print(f"❌ Không thể gọi hermes CLI để tạo {profile}: {e}")
            print(f"🛠️ Tiến hành tạo thư mục thủ công cho profile: {profile}")
            os.makedirs(profile_dir, exist_ok=True)
    else:
        print(f"ℹ️ Profile '{profile}' đã tồn tại.")

    if template_dir:
        soul_source = os.path.join(template_dir, f"soul-{profile}.md")
        if os.path.exists(soul_source):
            try:
                os.makedirs(profile_dir, exist_ok=True)
                shutil.copy(soul_source, soul_target)
                print(f"📝 Đã sao chép SOUL.md thành công cho '{profile}'")
            except Exception as e:
                print(f"❌ Lỗi khi sao chép SOUL.md cho '{profile}': {e}")
        else:
            print(f"⚠️ Không tìm thấy file nguồn: {soul_source}")

    # Đảm bảo config.yaml tồn tại trong profile
    if not os.path.exists(config_target):
        global_config = os.path.expanduser("~/.hermes/config.yaml")
        if os.path.exists(global_config):
            try:
                shutil.copy(global_config, config_target)
                print(f"⚙️ Đã sao chép config.yaml cấu hình toàn cục cho '{profile}'")
            except Exception as e:
                print(f"❌ Lỗi sao chép config.yaml cho '{profile}': {e}")
        else:
            # Tạo config.yaml mặc định
            default_config = "model:\n  default: qwen3.5:0.8b\n  provider: custom\n  base_url: http://localhost:11434/v1\nproviders: {}\n"
            try:
                with open(config_target, "w", encoding="utf-8") as f:
                    f.write(default_config)
                print(f"⚙️ Đã tự động tạo config.yaml mặc định cho '{profile}'")
            except Exception as e:
                print(f"❌ Lỗi khi tạo config.yaml mặc định cho '{profile}': {e}")

# 3. Tự động chuẩn bị thư mục làm việc output/vtn-session05 và file skill.json
project_dir = os.path.abspath(os.path.join(notebook_dir, "..", "..", "output", "vtn-session05"))
os.makedirs(project_dir, exist_ok=True)
print(f"📂 Thư mục làm việc của project: {project_dir}")

skill_target = os.path.join(project_dir, "skill.json")
if template_dir:
    skill_source = os.path.join(template_dir, "skill.json")
    if os.path.exists(skill_source):
        try:
            shutil.copy(skill_source, skill_target)
            print(f"📝 Đã chuẩn bị skill.json mặc định tại: {skill_target}")
        except Exception as e:
            print(f"❌ Lỗi khi sao chép skill.json: {e}")
    else:
        print(f"⚠️ Không tìm thấy file nguồn: {skill_source}")
```

```python
# Hiển thị nội dung SOUL.md cho từng profile
import os

profiles = ["tri-thuc-noi-bo", "soan-thao-noi-dung", "checklist-van-hanh"]
base = os.path.expanduser("~/.hermes/profiles")

for profile_name in profiles:
    soul_path = os.path.join(base, profile_name, "SOUL.md")
    print(f"{'='*60}")
    print(f"Profile: {profile_name}")
    print(f"{'='*60}")
    if os.path.exists(soul_path):
        with open(soul_path, "r", encoding="utf-8") as f:
            content = f.read()
        print(content[:500])
        if len(content) > 500:
            print(f"\n... (còn {len(content) - 500} ký tự)")
    else:
        print(f"❌ Chưa tìm thấy hoặc cấu hình sai: {soul_path}")
    print()
```

```python
# Hiển thị nội dung skill.json
import os
import json

notebook_dir = os.getcwd()
project_dir = os.path.abspath(os.path.join(notebook_dir, "..", "..", "output", "vtn-session05"))
skill_path = os.path.join(project_dir, "skill.json")

if os.path.exists(skill_path):
    with open(skill_path, "r", encoding="utf-8") as f:
        data = json.load(f)
    print("skill.json:")
    print(json.dumps(data, indent=2, ensure_ascii=False))
else:
    print(f"❌ Chưa tìm thấy skill.json tại: {skill_path}")
```
