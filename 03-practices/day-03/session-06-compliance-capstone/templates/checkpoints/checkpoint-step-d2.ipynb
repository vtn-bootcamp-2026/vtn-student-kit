### Bước 1: Khởi tạo thư mục `capstone_project/` và sao chép biểu mẫu
Chạy cell này để tự động tạo thư mục `capstone_project/` trong `session-06-compliance-capstone/` và sao chép toàn bộ 9 biểu mẫu hồ sơ từ `templates/` vào đó. Sau khi chạy xong, học viên mở các file trong thư mục đó và điền nội dung thực tế của nhóm.

```python
import shutil
from pathlib import Path

# Xác định thư mục session-06
current_dir = Path.cwd()
session_dir = None
for parent in [current_dir] + list(current_dir.parents):
    if parent.name == "session-06-compliance-capstone":
        session_dir = parent
        break
if not session_dir:
    session_dir = current_dir

templates_dir = session_dir / "templates"
capstone_dir = session_dir / "capstone_project"
capstone_dir.mkdir(parents=True, exist_ok=True)
print(f"[OK] Đã tạo thư mục capstone_project/ tại: {capstone_dir}")

# Danh sách 9 biểu mẫu cần sao chép
template_files = [
    "use-case-one-pager.md",
    "logical-workflow.md",
    "core-prompt-design.md",
    "compliance-checklist.md",
    "action-plan-30-90-days.md",
    "runbook-template.md",
    "handoff-contract.md",
    "failure-modes-rollback.md",
    "test-cases-specification.md",
]

copied = 0
print("\nSao chép biểu mẫu hồ sơ vào capstone_project/:")
for fname in template_files:
    src = templates_dir / fname
    dst = capstone_dir / fname
    if src.exists():
        if not dst.exists():
            shutil.copy(src, dst)
            print(f"  [OK] {fname}")
        else:
            print(f"  [ĐÃ CÓ] {fname} (giữ nguyên bản học viên đã chỉnh)")
        copied += 1
    else:
        print(f"  [THIẾU] Không tìm thấy {fname} trong templates/")

print(f"\n=> Đã sao chép {copied}/{len(template_files)} biểu mẫu vào capstone_project/.")
print("Học viên hãy mở thư mục capstone_project/ và điền nội dung vào các file trước khi đóng gói.")
```

### Bước 2: Học viên tự điền nội dung vào hồ sơ Capstone

> **⏸️ DỪNG LẠI — Học viên thực hiện thủ công trước khi chạy tiếp**

**YÊU CẦU QUAN TRỌNG:** Học viên bắt buộc phải mở thư mục `capstone_project/` và chỉnh sửa/bổ sung nội dung thực tế của nhóm vào **toàn bộ 9 tệp tin mẫu** dưới đây để hoàn thiện dự án Capstone. Thay thế toàn bộ các nhãn nháp/placeholder (dạng `[TODO]`, `[Điền tên nhóm]`, `[Điền ngày]`,...) bằng nội dung thực chất của dự án:

**Sau khi hoàn tất việc chỉnh sửa**, quay lại chạy Bước 3 để kiểm định.

### Bước 3: Tự động kiểm định chất lượng 9 hồ sơ Implementation Kit
Cell này quét 9 tài liệu trong `capstone_project/` để xác nhận học viên đã điền nội dung đầy đủ. Tiêu chí:
1. **Tồn tại**: File có mặt trong `capstone_project/`.
2. **Đã điền**: Dung lượng > 100 ký tự và KHÔNG còn nhãn placeholder `[TODO]`, `[DIEN_VAO]`, `[Điền tên nhóm]`, `[Điền ngày]`, v.v.

```python
required_docs = {
    "use-case-one-pager.md":       "Phiếu mô tả Use Case 1 trang",
    "logical-workflow.md":         "Sơ đồ luồng logic & Phân vai AI",
    "core-prompt-design.md":       "Đặc tả cấu trúc Prompt cốt lõi",
    "compliance-checklist.md":     "Bảng tự kiểm tra tuân thủ bảo mật",
    "action-plan-30-90-days.md":   "Lộ trình áp dụng 30-90 ngày",
    "runbook-template.md":         "Tài liệu hướng dẫn vận hành Runbook",
    "handoff-contract.md":         "Biên bản bàn giao sản phẩm kỹ thuật",
    "failure-modes-rollback.md":   "Phân tích lỗi & kịch bản Rollback",
    "test-cases-specification.md": "Đặc tả kịch bản ca kiểm thử",
}

PLACEHOLDER_KEYWORDS = [
    "[TODO]", "[DIEN_VAO]", "[ENTER_HERE]", 
    "[Điền tên nhóm]", "[Điền tên trưởng nhóm]", 
    "[Điền tên]", "[Điền ngày]", "[Điền tháng/năm]", 
    "[Ví dụ:"
]

print(f"{'Tên tệp tin':<32} | {'Mô tả tài liệu':<36} | Trạng thái")
print("-" * 95)

incomplete_count = 0

for fname, desc in required_docs.items():
    fpath = capstone_dir / fname
    if not fpath.exists():
        status = "🔴 THIẾU FILE (FAIL)"
        incomplete_count += 1
    else:
        content = fpath.read_text(encoding="utf-8").strip()
        has_placeholder = any(kw in content for kw in PLACEHOLDER_KEYWORDS)
        has_length = len(content) > 100
        if has_placeholder:
            status = "🟡 CHƯA XONG (còn placeholder)"
            incomplete_count += 1
        elif not has_length:
            status = "🟡 CHƯA ĐIỀN (nội dung quá ngắn)"
            incomplete_count += 1
        else:
            status = "🟢 HOÀN THÀNH (PASS)"
    print(f"{fname:<32} | {desc:<36} | {status}")

print("-" * 95)
if incomplete_count == 0:
    print("🎉 TUYỆT VỜI: 9/9 hồ sơ Implementation Kit đã hoàn thành. Sẵn sàng đóng gói!")
else:
    print(f"⚠️  CÒN {incomplete_count} hồ sơ chưa hoàn thiện.")
    print("   Mở các file trong capstone_project/ và điền đầy đủ nội dung trước khi đóng gói.")
```

### Bước 4: Đóng gói sản phẩm bàn giao (ZIP chéo nền tảng)
Chạy cell này **sau khi Bước 3 đạt 9/9** để tự động nén toàn bộ `capstone_project/` (9 hồ sơ) cùng với mã nguồn `src/anonymizer.py` và dữ liệu kiểm thử vào một tệp `.zip` bằng thư viện `zipfile` Python thuần (tương thích Windows & macOS).

Nhớ điền tên nhóm của bạn vào biến `ten_nhom` trong cell dưới!

```python
import zipfile

# ✏️ ĐIỀN TÊN NHÓM CỦA BẠN VÀO ĐÂY
ten_nhom = "Nhom-01"  # Ví dụ: Nhom-01, Nhom-Bao-Mat-VTN, ...

outputs_dir = session_dir / "outputs"
outputs_dir.mkdir(parents=True, exist_ok=True)
zip_filename = outputs_dir / f"capstone-handover-{ten_nhom}.zip"

# Nội dung đóng gói: capstone_project/ + src/ + synthetic-data/
paths_to_pack = [
    (capstone_dir,                "capstone_project"),
    (session_dir / "src",         "src"),
    (session_dir / "synthetic-data", "synthetic-data"),
]

print(f"Đang tạo gói bàn giao: {zip_filename.name}...")

try:
    with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for path, arcname in paths_to_pack:
            if path.exists() and path.is_dir():
                for file_path in path.rglob('*'):
                    if file_path.is_file() and '.ipynb_checkpoints' not in str(file_path):
                        rel_path = arcname + '/' + str(file_path.relative_to(path))
                        zipf.write(file_path, rel_path)
                print(f"  [Thêm thư mục] {arcname}/")
            elif path.exists() and path.is_file():
                zipf.write(path, arcname)
                print(f"  [Thêm tệp] {arcname}")
            else:
                print(f"  [Bỏ qua] {arcname} (không tồn tại)")

    size_mb = zip_filename.stat().st_size / (1024 * 1024)
    print(f"\n🎉 ĐÓNG GÓI THÀNH CÔNG!")
    print(f"  - Tên tệp nén : {zip_filename.name}")
    print(f"  - Vị trí lưu  : outputs/{zip_filename.name}")
    print(f"  - Kích thước   : {size_mb:.2f} MB")
    print(f"\n👉 Học viên vui lòng nộp tệp tin .zip này lên hệ thống để hoàn tất bài thực hành Capstone!")
except Exception as e:
    print(f"❌ Đóng gói thất bại: {e}")
```
