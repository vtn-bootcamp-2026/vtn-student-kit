# Checkpoint C: Khởi tạo thư mục bài tập nhóm

Checkpoint này giúp tự động hóa việc chuẩn bị thư mục làm việc cho bài tập nhóm ở **Phần C: Bài tập nhóm - chọn và chấm bài toán**.

### Các công việc notebook này thực hiện:
1. Tạo thư mục `outputs/bai_tap_nhom/` nếu chưa tồn tại.
2. Sao chép các tệp template tương ứng với Phần C từ thư mục `templates/` sang `outputs/bai_tap_nhom/`:
   - `use-case-one-pager.md` (Mẫu bản mô tả bài toán một trang)
   - `use-case-scoring-rubbric.md` (Mẫu thang chấm điểm lựa chọn bài toán)
   - `risk-control-checklist.md` (Mẫu bảng kiểm rủi ro sơ bộ)
   - `use-case-prompt.md` (Mẫu lời nhắc đồng sáng tạo use case cùng AI)

### Hướng dẫn sử dụng:
Nhấn biểu tượng **Play** (hoặc nhấn tổ hợp phím `Shift + Enter`) ở cell code Python bên dưới để bắt đầu.

```python
import os
import shutil

# Thiết lập các đường dẫn tương đối
# Thư mục hiện tại của notebook: templates/checkpoint/
TEMPLATE_DIR = ".."
OUTPUT_DIR = "../../outputs/bai_tap_nhom"

print("=== Bắt đầu khởi tạo thư mục bài tập nhóm ===\n")

# 1. Tạo thư mục outputs/bai_tap_nhom/
try:
    os.makedirs(OUTPUT_DIR, exist_ok=True)
    print(f"[OK] Đã xác nhận/tạo thư mục đích: {OUTPUT_DIR}")
except Exception as e:
    print(f"[FAIL] Lỗi khi tạo thư mục: {e}")

# 2. Danh sách các file template cần sao chép cho phần C
files_to_copy = [
    "use-case-one-pager.md",
    "use-case-scoring-rubbric.md",
    "risk-control-checklist.md",
    "use-case-prompt.md"
]

print("\n--- Sao chép các tệp mẫu ---")
for filename in files_to_copy:
    src_path = os.path.join(TEMPLATE_DIR, filename)
    dst_path = os.path.join(OUTPUT_DIR, filename)
    
    if os.path.exists(src_path):
        try:
            shutil.copy2(src_path, dst_path)
            print(f"  [OK] Đã sao chép: {filename} -> {dst_path}")
        except Exception as e:
            print(f"  [FAIL] Lỗi khi sao chép {filename}: {e}")
    else:
        print(f"  [FAIL] Không tìm thấy tệp mẫu nguồn: {src_path}")

print("\n=== Hoàn thành chuẩn bị dữ liệu! ===")
```

### Kết quả kỳ vọng:
- Thư mục `outputs/bai_tap_nhom/` đã được tạo thành công.
- 4 file mẫu đã xuất hiện đầy đủ trong thư mục này.
