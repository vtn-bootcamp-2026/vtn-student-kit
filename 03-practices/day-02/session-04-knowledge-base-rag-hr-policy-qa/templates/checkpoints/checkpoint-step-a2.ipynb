# Checkpoint A2: SKILL.md va skill.json hoan chinh

Day la vi du hoan chinh cua 2 file ban can tao.

**Cach su dung:** Chay tat ca cells. So sanh ket qua voi file ban da tao.

```python
import json
import re
from pathlib import Path

SKILL_DIR = Path("../../templates/skills/hr-policy-qa-skill")
skill_md_path = SKILL_DIR / "SKILL.md"

if not skill_md_path.exists():
    print(f"[LOI] Khong tim thay SKILL.md tai: {skill_md_path}")
    print("Hay kiem tra duong dan tu thu muc lam viec hien tai.")
else:
    with open(skill_md_path, "r", encoding="utf-8") as f:
        skill_md_content = f.read()
    print(f"Da doc SKILL.md: {len(skill_md_content)} ky tu, {skill_md_content.count(chr(10))} dong")
```

```python
# Phan tich 6 sections cua SKILL.md
# SKILL.md co cau truc: ## 1. ... den ## 6. ...

section_pattern = r"^## \d+\.\s+(.+)$"
section_headers = re.findall(section_pattern, skill_md_content, re.MULTILINE)

# Tach noi dung tung section
sections = re.split(r"(?=^## \d+\.)", skill_md_content, flags=re.MULTILINE)
# Bo phan dau tien (frontmatter + header chung)
sections = [s for s in sections if s.strip().startswith("## ")]

print(f"SKILL.md co {len(sections)} sections:\n")

for i, section in enumerate(sections, 1):
    # Lay tieu de section
    first_line = section.strip().split("\n")[0]
    title = re.sub(r"^## \d+\.\s+", "", first_line)
    line_count = section.count("\n")
    word_count = len(section.split())
    
    print(f"--- Section {i}: {title} ---")
    print(f"    Dong: {line_count} | Tu: {word_count}")
    # Hien thi 5 dong dau cua section
    preview_lines = section.strip().split("\n")[:5]
    for line in preview_lines:
        print(f"    {line[:80]}")
    print()
```

```python
# Doc va validate skill.json
skill_json_path = SKILL_DIR / "skill.json"

with open(skill_json_path, "r", encoding="utf-8") as f:
    skill_json = json.load(f)

# Kiem tra cac truong bat buoc
required_fields = ["name", "version", "triggers", "permissions", "quality_gates"]
print("Validation skill.json:")
print("=" * 50)

all_present = True
for field in required_fields:
    present = field in skill_json
    status = "\u2713" if present else "\u2717"
    print(f"  {status} {field}: {'CO' if present else 'THIEU'}")
    if present:
        val = skill_json[field]
        if isinstance(val, dict):
            print(f"      -> {len(val)} sub-keys: {', '.join(list(val.keys())[:5])}")
        elif isinstance(val, list):
            print(f"      -> {len(val)} items")
        else:
            print(f"      -> {val}")
    if not present:
        all_present = False

print(f"\nKet qua: {'HOP LE - Tat ca truong bat buoc deu co' if all_present else 'THIEU TRUONG!'}")

# Hien thi noi dung day du
print("\nNoi dung skill.json:")
print("=" * 50)
print(json.dumps(skill_json, indent=2, ensure_ascii=False))
```

```python
# In quality_gates bang formatted table
quality_gates = skill_json.get("quality_gates", {})

print("QUALITY GATES (skill.json):")
print("=" * 70)
print(f"{'Metric':<40} {'Target':<30}")
print("-" * 70)

for metric, target in quality_gates.items():
    print(f"{metric:<40} {target:<30}")

print("=" * 70)
print(f"\nTong so quality_gates: {len(quality_gates)}")
```

```python
# Tong ket cuoi cung
section_count = len(sections)
qg_count = len(skill_json.get("quality_gates", {}))

print("=" * 60)
print("TONG KET CHECKPOINT A2")
print("=" * 60)
print(f"SKILL.md co {section_count} sections \u2713")
print(f"skill.json co {qg_count} quality_gates \u2713")
print()
print("So sanh voi file cua ban:")
print("  - SKILL.md phai co du 6 sections:")
print("    1. Mo ta & Vai tro (Persona)")
print("    2. Kich ban kich hoat (Triggers)")
print("    3. Quy trinh thuc thi (Execution Workflow)")
print("    4. Dinh dang dau ra (Output Format)")
print("    5. Gioi han (Boundaries)")
print("    6. Quy tac an toan (Safety Rules)")
print("  - skill.json phai co: name, version, triggers, permissions, quality_gates")
print()
print("Checkpoint A2 hoan thanh! Tiep tuc voi checkpoint-step-b2.ipynb.")
```
