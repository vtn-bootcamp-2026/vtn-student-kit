# Checkpoint A2: SKILL.md và skill.json hoàn chỉnh

Đây là ví dụ hoàn chỉnh của 2 file bạn cần tạo.

**Cách sử dụng:** Chạy tất cả cells. So sánh kết quả với file bạn đã tạo.

```python
import json
import re
import shutil
from pathlib import Path

TEMPLATE_SKILL_DIR = Path("../../templates/skills/hr-policy-qa-skill")
SKILL_DIR = Path("../../outputs/skills/hr-policy-qa-skill")

# Tự động sao chép SKILL.md từ templates sang outputs nếu chưa tồn tại
src_skill_md = TEMPLATE_SKILL_DIR / "SKILL.md"
dst_skill_md = SKILL_DIR / "SKILL.md"
if not dst_skill_md.exists() and src_skill_md.exists():
    shutil.copy(src_skill_md, dst_skill_md)
    print(f"✓ Đã tạo SKILL.md tại: {dst_skill_md}")
else:
    print(f"SKILL.md đã tồn tại tại: {dst_skill_md}")

if not dst_skill_md.exists():
    print(f"[LỖI] Không tìm thấy SKILL.md tại: {dst_skill_md}")
else:
    with open(dst_skill_md, "r", encoding="utf-8") as f:
        skill_md_content = f.read()
    print(f"Đã đọc SKILL.md học viên: {len(skill_md_content)} ký tự, {skill_md_content.count(chr(10))} dòng")
```

```python
# Phân tích 6 sections của SKILL.md
# SKILL.md có cấu trúc: ## 1. ... đến ## 6. ...

section_pattern = r"^## \d+\.\s+(.+)$"
section_headers = re.findall(section_pattern, skill_md_content, re.MULTILINE)

# Tách nội dung từng section
sections = re.split(r"(?=^## \d+\.)", skill_md_content, flags=re.MULTILINE)
# Bỏ phần đầu tiên (frontmatter + header chung)
sections = [s for s in sections if s.strip().startswith("## ")]

print(f"SKILL.md có {len(sections)} sections:\n")

for i, section in enumerate(sections, 1):
    # Lấy tiêu đề section
    first_line = section.strip().split("\n")[0]
    title = re.sub(r"^## \d+\.\s+", "", first_line)
    line_count = section.count("\n")
    word_count = len(section.split())
    
    print(f"--- Section {i}: {title} ---")
    print(f"    Dòng: {line_count} | Từ: {word_count}")
    # Hiển thị 5 dòng đầu của section
    preview_lines = section.strip().split("\n")[:5]
    for line in preview_lines:
        print(f"    {line[:80]}")
    print()
```

```python
# Đọc và validate skill.json của học viên
src_skill_json = TEMPLATE_SKILL_DIR / "skill.json"
skill_json_path = SKILL_DIR / "skill.json"

# Tự động sao chép skill.json sang outputs nếu chưa tồn tại
if not skill_json_path.exists() and src_skill_json.exists():
    shutil.copy(src_skill_json, skill_json_path)
    print(f"✓ Đã tạo skill.json tại: {skill_json_path}")
else:
    print(f"skill.json đã tồn tại tại: {skill_json_path}")

with open(skill_json_path, "r", encoding="utf-8") as f:
    skill_json = json.load(f)

# Kiểm tra các trường bắt buộc
required_fields = ["name", "version", "triggers", "permissions", "quality_gates"]
print("Validation skill.json học viên:")
print("=" * 50)

all_present = True
for field in required_fields:
    present = field in skill_json
    status = "\u2713" if present else "\u2717"
    print(f"  {status} {field}: {'CÓ' if present else 'THIẾU'}")
    if present:
        val = skill_json[field]
        if isinstance(val, dict):
            print(f"      -> {len(val)} sub-keys: {', '.join(list(val.keys())[:5])}")
        elif isinstance(val, list):
            print(f"      -> {len(val)} items")
        else:
            print(f"      -> {val}")
    if not present:
        all_present = False

print(f"\nKết quả: {'HỢP LỆ - Tất cả trường bắt buộc đều có' if all_present else 'THIẾU TRƯỜNG!'}")

# Hiển thị nội dung đầy đủ
print("\nNội dung skill.json:")
print("=" * 50)
print(json.dumps(skill_json, indent=2, ensure_ascii=False))
```

```python
# In quality_gates bằng formatted table
quality_gates = skill_json.get("quality_gates", {})

print("QUALITY GATES (skill.json):")
print("=" * 70)
print(f"{'Metric':<40} {'Target':<30}")
print("-" * 70)

for metric, target in quality_gates.items():
    print(f"{metric:<40} {target:<30}")

print("=" * 70)
print(f"\nTổng số quality_gates: {len(quality_gates)}")
```

```python
# Tổng kết cuối cùng
section_count = len(sections)
qg_count = len(skill_json.get("quality_gates", {}))

print("=" * 60)
print("TỔNG KẾT CHECKPOINT A2")
print("=" * 60)
print(f"SKILL.md có {section_count} sections \u2713")
print(f"skill.json có {qg_count} quality_gates \u2713")
print()
print("So sánh với file của bạn:")
print("  - SKILL.md phải có đủ 6 sections:")
print("    1. Mô tả & Vai trò (Persona)")
print("    2. Kịch bản kích hoạt (Triggers)")
print("    3. Quy trình thực thi (Execution Workflow)")
print("    4. Định dạng đầu ra (Output Format)")
print("    5. Giới hạn (Boundaries)")
print("    6. Quy tắc an toàn (Safety Rules)")
print("  - skill.json phải có: name, version, triggers, permissions, quality_gates")
print()
print("Checkpoint A2 hoàn thành! Các file SKILL.md và skill.json đã được tạo tại thư mục outputs. Hãy tiếp tục với checkpoint-step-b2.ipynb.")
```
