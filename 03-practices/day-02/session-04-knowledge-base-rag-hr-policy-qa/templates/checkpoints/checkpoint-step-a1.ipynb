# Checkpoint A1: Phân tích Skill package mẫu

Sau khi chạy notebook này, bạn đã hiểu rõ cấu trúc Skill package.

**Cách sử dụng:** Chạy từng cell (Shift+Enter) hoặc chạy tất cả (Run All). Nếu thành công, bạn có thể tiếp tục từ Step A2.

```python
import json
import os
from pathlib import Path

print("Imports thành công!")
print(f"Python working dir: {os.getcwd()}")
```

```python
# Định nghĩa đường dẫn đến Skill package mẫu
SKILL_DIR = Path("../../templates/skills/hr-policy-qa-skill").resolve()

def print_tree(directory, prefix="", max_depth=5, current_depth=0):
    """In cấu trúc tree của thư mục."""
    if current_depth >= max_depth:
        return
    path = Path(directory)
    if not path.exists():
        print(f"[LỖI] Thư mục không tồn tại: {path}")
        return
    
    items = sorted(path.iterdir(), key=lambda x: (not x.is_dir(), x.name))
    for i, item in enumerate(items):
        is_last = (i == len(items) - 1)
        connector = "\u2514\u2500\u2500 " if is_last else "\u251c\u2500\u2500 "
        icon = "\U0001f4c1 " if item.is_dir() else "\U0001f4c4 "
        print(f"{prefix}{connector}{icon}{item.name}")
        if item.is_dir():
            extension = "    " if is_last else "\u2502   "
            print_tree(item, prefix + extension, max_depth, current_depth + 1)

print("Cấu trúc Skill package mẫu (hr-policy-qa-skill):")
print("=" * 50)
print_tree(SKILL_DIR)
```

```python
# Đọc và hiển thị SKILL.md (50 dòng đầu tiên)
skill_md_path = SKILL_DIR / "SKILL.md"

if skill_md_path.exists():
    with open(skill_md_path, "r", encoding="utf-8") as f:
        lines = f.readlines()
    
    print(f"SKILL.md ({len(lines)} dòng total) — Hiển thị 50 dòng đầu:")
    print("=" * 60)
    for i, line in enumerate(lines[:50], 1):
        print(f"{i:3d} | {line.rstrip()}")
    if len(lines) > 50:
        print(f"\n... (còn {len(lines) - 50} dòng nữa)")
else:
    print(f"[LỖI] Không tìm thấy: {skill_md_path}")
```

```python
# Đọc và hiển thị skill.json (formatted JSON)
skill_json_path = SKILL_DIR / "skill.json"

if skill_json_path.exists():
    with open(skill_json_path, "r", encoding="utf-8") as f:
        skill_data = json.load(f)
    
    print("skill.json — Nội dung hoàn chỉnh:")
    print("=" * 60)
    print(json.dumps(skill_data, indent=2, ensure_ascii=False))
else:
    print(f"[LỖI] Không tìm thấy: {skill_json_path}")
```

```python
# Đọc và hiển thị schema hr-response.schema.json
schema_path = SKILL_DIR / "schemas" / "hr-response.schema.json"

if schema_path.exists():
    with open(schema_path, "r", encoding="utf-8") as f:
        schema_data = json.load(f)
    
    print("schemas/hr-response.schema.json — Nội dung hoàn chỉnh:")
    print("=" * 60)
    print(json.dumps(schema_data, indent=2, ensure_ascii=False))
else:
    print(f"[LỖI] Không tìm thấy: {schema_path}")
```

```python
# Liệt kê scripts/ và in 30 dòng đầu mỗi file
scripts_dir = SKILL_DIR / "scripts"

if scripts_dir.exists():
    script_files = sorted(scripts_dir.glob("*.py"))
    print(f"scripts/ — {len(script_files)} file(s):\n")
    
    for script_file in script_files:
        with open(script_file, "r", encoding="utf-8") as f:
            lines = f.readlines()
        
        print(f"\u2550" * 60)
        print(f"FILE: scripts/{script_file.name} ({len(lines)} dòng)")
        print(f"\u2550" * 60)
        for i, line in enumerate(lines[:30], 1):
            print(f"{i:3d} | {line.rstrip()}")
        if len(lines) > 30:
            print(f"    ... (còn {len(lines) - 30} dòng nữa)")
        print()
else:
    print(f"[LỖI] Thư mục scripts/ không tồn tại")
```

```python
# Bảng tổng hợp: file -> mục đích -> số lượng sections/fields
print("BẢNG TỔNG HỢP SKILL PACKAGE")
print("=" * 80)
print(f"{'File':<45} {'Mục đích':<25} {'Details':<15}")
print("-" * 80)

# SKILL.md
skill_md_path = SKILL_DIR / "SKILL.md"
if skill_md_path.exists():
    with open(skill_md_path, "r", encoding="utf-8") as f:
        content = f.read()
    h2_count = content.count("\n## ")
    print(f"  SKILL.md                                  Hướng dẫn kỹ năng       {h2_count} sections")

# skill.json
skill_json_path = SKILL_DIR / "skill.json"
if skill_json_path.exists():
    with open(skill_json_path, "r", encoding="utf-8") as f:
        data = json.load(f)
    top_keys = len(data)
    qg_count = len(data.get("quality_gates", {}))
    print(f"  skill.json                                Cấu hình skill          {top_keys} fields, {qg_count} quality_gates")

# schema
schema_path = SKILL_DIR / "schemas" / "hr-response.schema.json"
if schema_path.exists():
    with open(schema_path, "r", encoding="utf-8") as f:
        data = json.load(f)
    required_count = len(data.get("required", []))
    props_count = len(data.get("properties", {}))
    print(f"  schemas/hr-response.schema.json           JSON response schema    {required_count} required, {props_count} properties")

# scripts
scripts_dir = SKILL_DIR / "scripts"
if scripts_dir.exists():
    for sf in sorted(scripts_dir.glob("*.py")):
        with open(sf, "r", encoding="utf-8") as f:
            lines = f.readlines()
        funcs = [l.strip() for l in lines if l.strip().startswith("def ")]
        print(f"  scripts/{sf.name:<35} Python script          {len(funcs)} functions")

# kb files
kb_dir = SKILL_DIR / "kb" / "hr-policies"
if kb_dir.exists():
    kb_files = list(kb_dir.glob("*.md"))
    print(f"  kb/hr-policies/                           Tài liệu chính sách     {len(kb_files)} files")

# kb-inventory
inv_path = SKILL_DIR / "kb" / "kb-inventory.md"
if inv_path.exists():
    print(f"  kb/kb-inventory.md                        Danh mục kho tri thức   1 file")

print("=" * 80)
print("\nCheckpoint A1 hoàn thành! Bạn đã hiểu rõ cấu trúc Skill package.")
print("Hãy chạy các cell tiếp theo bên dưới để khởi tạo cấu trúc thư mục Skill package của riêng nhóm bạn.")
```

### Khởi tạo thư mục Skill package của riêng nhóm bạn

Bây giờ, chúng ta sẽ bắt đầu xây dựng Skill package của riêng mình bằng cách tạo cấu trúc thư mục trống dưới đường dẫn `outputs/skills/hr-policy-qa-skill`.
Thư mục này sẽ lưu trữ các file `SKILL.md`, `skill.json`, các tài liệu tri thức `kb/`, schema và các scripts xử lý logic.

```python
# Khởi tạo cấu trúc thư mục Skill package cho học viên tại outputs
OUTPUT_SKILL_DIR = Path("../../outputs/skills/hr-policy-qa-skill").resolve()
subdirs = [
    OUTPUT_SKILL_DIR / "kb" / "hr-policies",
    OUTPUT_SKILL_DIR / "schemas",
    OUTPUT_SKILL_DIR / "scripts"
]

print("Đang khởi tạo cấu trúc thư mục Skill package tại outputs...")
for subdir in subdirs:
    subdir.mkdir(parents=True, exist_ok=True)
    print(f"  ✓ Đã tạo: {subdir.relative_to(OUTPUT_SKILL_DIR.parent.parent.parent)}")

print("\nCấu trúc thư mục hiện tại của outputs/skills/hr-policy-qa-skill:")
print("=" * 60)
print_tree(OUTPUT_SKILL_DIR)

print("\nKhởi tạo thành công! Hãy chuyển sang checkpoint-step-a2.ipynb để tạo SKILL.md và skill.json.")
```
