# Checkpoint A2: Viết SKILL.md — bản đồ chỉ dẫn cho Agent

Checkpoint này tạo SKILL.md — trái tim của Agent Skill.

**SKILL.md cần có 6 phần:**
1. Persona — vai trò và nguyên tắc cốt lõi
2. Triggers — khi nào Agent kích hoạt Skill
3. Execution workflow — 4 bước: intake → extraction → self-check → routing
4. Output format — tham chiếu JSON schema, các trường bắt buộc
5. Boundaries — ranh giới xử lý
6. Safety rules — quy tắc an toàn

**Điều kiện đạt:**
- [x] SKILL.md có đủ 6 phần
- [x] Workflow mô tả rõ 4 bước, mỗi bước nêu script cần gọi
- [x] Có ví dụ lệnh cụ thể

**Tiếp theo:** Chạy cell bên dưới để tạo SKILL.md mẫu, sau đó chỉnh sửa cho phù hợp. Chuyển sang Checkpoint A3 (skill.json).

```python
import os
import shutil

PROJECT_ROOT = "../../outputs/contract-term-extractor"
src_skill = "../../templates/skills/contract-term-extractor/SKILL.md"
dst_skill = f"{PROJECT_ROOT}/SKILL.md"

# Copy file SKILL.md từ template chuẩn
os.makedirs(os.path.dirname(dst_skill), exist_ok=True)
shutil.copy(src_skill, dst_skill)
print(f"Đã tạo SKILL.md tại: {dst_skill}")

with open(dst_skill, "r", encoding="utf-8") as f:
    skill_md = f.read()

print(f"Kích thước: {len(skill_md)} ký tự")
print()
print("Nội dung SKILL.md:")
print("=" * 50)
print(skill_md)
```
