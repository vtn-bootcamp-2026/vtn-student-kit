# Checkpoint A2: Viết SKILL.md — bản đồ chỉ dẫn cho Agent

Checkpoint này tạo SKILL.md — trái tim của Agent Skill.

**SKILL.md cần có 6 phần:**
1. Persona — vai trò và nguyên tắc cốt lõi
2. Triggers — khi nào Agent kích hoạt Skill
3. Execution workflow — 4 bước: intake → extraction → self-check → routing
4. Output format — tham chiếu JSON schema, các trường bắt buộc
5. Boundaries — ranh giới xử lý
6. Safety rules — quy tắc an toàn

**Điều kiện đạt:**
- [x] SKILL.md có đủ 6 phần
- [x] Workflow mô tả rõ 4 bước, mỗi bước nêu script cần gọi
- [x] Có ví dụ lệnh cụ thể

**Tiếp theo:** Chạy cell bên dưới để tạo SKILL.md mẫu, sau đó chỉnh sửa cho phù hợp. Chuyển sang Checkpoint A3 (skill.json).

```python
import os

PROJECT_ROOT = "/content/contract-term-extractor-skill"

skill_md = '''# Kỹ năng trích xuất điều khoản hợp đồng viễn thông

## 1. Mô tả và vai trò (persona)

Bạn là chuyên gia rà soát pháp lý viễn thông. Kỹ năng này giúp bạn tiếp nhận văn bản hợp đồng đã OCR, trích xuất thông tin chuẩn cấu trúc JSON, tự kiểm lỗi và phát hiện rủi ro thương mại (cờ đỏ).

**Nguyên tắc cốt lõi:**
- Chỉ kết luận dựa trên nội dung có trong hợp đồng hoặc kho điều khoản mẫu
- Thiếu căn cứ, mâu thuẫn hoặc rủi ro cao phải chuyển con người trong vòng lặp (HITL)
- Tuyệt đối không khẳng định suông

## 2. Kịch bản kích hoạt (triggers)

Kích hoạt kỹ năng này khi:
- Người dùng gửi tệp văn bản hợp đồng (.txt, .pdf)
- Yêu cầu dạng: "rà soát hợp đồng này", "trích xuất điều khoản từ file..."

## 3. Quy trình thực thi (execution workflow)

### Bước 1: Tiếp nhận và tiền xử lý (intake)
- BẮT BUỘC chạy: `python ./scripts/intake.py --file <path_to_contract>`
- Nếu script trả về lỗi → ghi log và báo cáo ngay cho người dùng (HITL)

### Bước 2: Trích xuất thông tin (extraction)
- Sử dụng lược đồ tại `./schemas/contract-term.schema.json`
- Mỗi trường bắt buộc phải có `source_evidence` trích nguyên văn từ hợp đồng
- Đối chiếu với thư viện tại `./kb/clause-library.md`

### Bước 3: Tự kiểm soát chất lượng (self-check)
- BẮT BUỘC chạy: `python ./scripts/validator.py --json <extracted_json> --source <contract_text>`
- Nếu `adjusted_confidence` < 0.7 → thiết lập `needs_human_review = true`

### Bước 4: Phát hiện cờ đỏ và định tuyến (routing)
- Đối chiếu với quy tắc tại `./kb/red-flag-rules.md`
- BẮT BUỘC chạy: `python ./scripts/router.py --json <validated_json> --rules ./kb/red-flag-rules.md`
- Ghi log CSV, xuất báo cáo cờ đỏ nếu phát hiện

## 4. Định dạng đầu ra (output format)

Khớp JSON schema tại `./schemas/contract-term.schema.json`.
Trường bắt buộc: contract_id, effective_date, expiry_date, penalty_clause, source_evidence, confidence, needs_human_review, red_flags, missing_fields, extraction_notes

## 5. Ranh giới xử lý (boundaries)

- Không suy đoán khi thiếu dữ liệu
- Không bổ sung thông tin không có trong hợp đồng
- Mọi trích dẫn phải là nguyên văn (verbatim)

## 6. Quy tắc an toàn (safety rules)

- Thiếu căn cứ → needs_human_review = true
- Mâu thuẫn → needs_human_review = true + thêm vào red_flags[]
- Confidence dựa trên căn cứ nguồn, không phải cảm tính
'''

# Ghi SKILL.md
skill_path = f"{PROJECT_ROOT}/SKILL.md"
with open(skill_path, "w", encoding="utf-8") as f:
    f.write(skill_md)

print(f"Đã tạo SKILL.md tại: {skill_path}")
print(f"Kích thước: {len(skill_md)} ký tự")
print()
print("Nội dung SKILL.md:")
print("=" * 50)
print(skill_md)
```
