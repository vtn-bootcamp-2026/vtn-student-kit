# Checkpoint A1: Tạo cấu trúc thư mục Agent Skill

Checkpoint này tạo cấu trúc thư mục `contract-term-extractor/` theo chuẩn Agent Skill.

**Cấu trúc cần tạo:**
```
contract-term-extractor/
  SKILL.md                 ← Bản đồ chỉ dẫn cho Agent
  skill.json               ← Metadata, triggers, permissions
  schemas/
  kb/
  scripts/
  references/
  data/contracts/
  outputs/
  tests/
```

**Điều kiện đạt:**
- [x] Tạo đủ thư mục con
- [x] Copy hoặc tạo tệp dữ liệu tĩnh từ template chuẩn
- [x] Copy 4 hợp đồng mô phỏng thực tế (.docx) và chỉ mục của chúng từ synthetic-data/ vào data/
- [x] Phân vai nhóm rõ ràng

**Tiếp theo:** Chạy cell bên dưới, sau đó chuyển sang Checkpoint A2 (viết SKILL.md).

````python
import os
import shutil

# Cấu hình đường dẫn — sử dụng relative path trỏ vào thư mục outputs
PROJECT_ROOT = "../../outputs/contract-term-extractor"

# Dọn dẹp thư mục cũ để hỗ trợ chạy thử độc lập nhiều lần
if os.path.exists(PROJECT_ROOT):
    shutil.rmtree(PROJECT_ROOT)
    print(f"Đã dọn dẹp thư mục cũ tại: {PROJECT_ROOT}")

# Tạo cấu trúc thư mục Agent Skill
dirs = [
    PROJECT_ROOT,
    f"{PROJECT_ROOT}/schemas",
    f"{PROJECT_ROOT}/kb",
    f"{PROJECT_ROOT}/scripts",
    f"{PROJECT_ROOT}/references",
    f"{PROJECT_ROOT}/data/contracts",
    f"{PROJECT_ROOT}/outputs/extracted-terms",
    f"{PROJECT_ROOT}/outputs/reports",
    f"{PROJECT_ROOT}/tests",
]

for d in dirs:
    os.makedirs(d, exist_ok=True)

print(f"Đã tạo cấu trúc Agent Skill tại: {PROJECT_ROOT}")
print()

# Kiểm tra
for d in dirs:
    exists = "OK" if os.path.exists(d) else "MISSING"
    print(f"  [{exists}] {d}")
````

````python
# ============================================================
# Sao chép các tệp cấu hình và tri thức chuẩn từ template gốc
# ============================================================
import shutil
import os

TEMPLATE_DIR = "../../templates/skills/contract-term-extractor"

print("Bắt đầu sao chép các tệp chuẩn sang project...")

# 1. Sao chép skill.json
if os.path.exists(f"{TEMPLATE_DIR}/skill.json"):
    shutil.copy(f"{TEMPLATE_DIR}/skill.json", f"{PROJECT_ROOT}/skill.json")
    print("  [OK] Đã sao chép skill.json")

# 2. Sao chép các thư mục schemas, kb, references
for folder in ["schemas", "kb", "references"]:
    src_folder = f"{TEMPLATE_DIR}/{folder}"
    dst_folder = f"{PROJECT_ROOT}/{folder}"
    if os.path.exists(src_folder):
        if os.path.exists(dst_folder):
            shutil.rmtree(dst_folder)
        shutil.copytree(src_folder, dst_folder)
        print(f"  [OK] Đã sao chép thư mục {folder}/")

print("\nHoàn thành chuẩn bị dữ liệu tĩnh!")
````

````python
# Copy hợp đồng mô phỏng thực tế từ synthetic-data/
import shutil
import os

SYNTHETIC_DIR = "../../synthetic-data"

print("Sao chép các hợp đồng mô phỏng thực tế từ synthetic-data/...")

# 1. Sao chép contracts-index.csv
if os.path.exists(f"{SYNTHETIC_DIR}/contracts-index.csv"):
    shutil.copy(f"{SYNTHETIC_DIR}/contracts-index.csv", f"{PROJECT_ROOT}/data/contracts-index.csv")
    print("  [OK] Đã sao chép contracts-index.csv")

# 2. Sao chép thư mục contracts (.docx)
src_contracts = f"{SYNTHETIC_DIR}/contracts"
dst_contracts = f"{PROJECT_ROOT}/data/contracts"

if os.path.exists(src_contracts):
    if os.path.exists(dst_contracts):
        shutil.rmtree(dst_contracts)
    shutil.copytree(src_contracts, dst_contracts)
    print("  [OK] Đã sao chép thư mục contracts chứa các file .docx")
    
    # Liệt kê các file đã copy
    print("\nCác tệp hợp đồng đã sao chép:")
    for fname in os.listdir(dst_contracts):
        if not fname.startswith("."):
            print(f"  - {fname}")
else:
    print("  [FAIL] Không tìm thấy thư mục nguồn contracts tại", src_contracts)
````
